```python
movies= pd.read_csv("datasets/movies_transformed.csv", index_col=[0])
```

```python
bins = [0, 2, 4, 6, 8, 10]
labels = ['0-2', '3-4', '5-6', '7-8', '9-10']
vote_range = {"votes" : movies["vote_average"].values}
df = pd.DataFrame(vote_range)

df["vote_range"] =pd.cut(df["votes"], bins=bins, labels=labels, right=False)
plt.figure(figsize=(10, 6))
sns.countplot(x = 'vote_range', data = df, palette='viridis')
plt.title("Count of movies by vote_average")
plt.xlabel("Vote Range")
plt.ylabel("Count of Movies")
plt.show()

```

```python
Popularity = {"popularity" : movies["popularity"].values}
df = pd.DataFrame(Popularity)
df.sort_values(by=["popularity"])

plt.figure(figsize = (10,10))
plt.title("Most Popular Movies")
sns.barplot(x = 'popularity', y = 'title', data = movies.sort_values(by=["popularity"]).tail(10), palette="magma")
```

```python
years = movies["release_year"].values

years_series = pd.Series(years)

year_counts = years_series.value_counts()

top_10_years = year_counts.head(10)
print(top_10_years)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_years.index, y=top_10_years.values, palette='viridis')
plt.title("Top 10 Years by Debuts (10 years straight)")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
titles = movies["title"].values

words = " ".join(titles)

wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(words)  #no parece encontrar wordcloud

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  
plt.title("Word Cloud of Titles")
plt.show()
```
